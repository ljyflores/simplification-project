```python
import argparse
from datasets import load_dataset
from utils_eval import compute_metrics

# Get dataset from arguments

DATASET_NAME = "cnn_full"

dataset = load_dataset(
    "json", data_files=f"data/{DATASET_NAME}_multiple.json", field="train"
)
dataset["test"] = load_dataset(
    "json", data_files=f"data/{DATASET_NAME}_multiple.json", field="test"
)["train"]

sources = [item["input"] for item in dataset["test"]]
labels = [item["labels"] for item in dataset["test"]]

for preds_path in [
    # "output/cochrane_full_bart_xsum_lt_1ep.txt"
    # "output/sota/in_domain/xsum_full_bart_xsum_no_finetune.txt",
    # "output/sota/in_domain/xsum_full_bart_xsum_finetune1ep.txt",
    # "output/mi/xsum_full_bart_xsum_lt_1ep.txt",
    # "output/clean/xsum_full_bart_xsum_drop_ex_1ep.txt",
    # "output/clean/xsum_full_bart_xsum_drop_sent_1ep.txt"
    "output/cnn_full_bart_xsum_lt_1e8_04.txt"
]:
    with open(preds_path) as f:
        preds = f.read().splitlines()

    if len(preds) > len(sources):
        preds = list(filter(lambda s: s != "", preds))

    result = compute_metrics(
        sources,
        preds,
        labels,
        ["check_entities"],
    )
    print(result)
```
