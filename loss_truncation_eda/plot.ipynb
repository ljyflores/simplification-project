```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
label = {-1: "Non-Factual", 1: "Factual"}
```

```python
df_ep0 = pd.read_csv("experiment_xsum_trained.csv")
df_ep1 = pd.read_csv("experiment_xsum_trained_finetune.csv")
```

```python
df_ep0["nli_flag_ent_output"] = df_ep0["nli_flag_ent_output"].apply(lambda x: label[x])
df_ep1["nli_flag_ent_output"] = df_ep1["nli_flag_ent_output"].apply(lambda x: label[x])
```

```python
sns.distplot(
    a = df_ep0.loc[df_ep0["nli_flag_ent_output"] == "Non-Factual"]["nll_mean"], 
    hist=False,
    kde_kws=dict(linewidth=3)
    )
sns.distplot(
    a = df_ep0.loc[df_ep0["nli_flag_ent_output"] == "Factual"]["nll_mean"], 
    hist=False,
    kde_kws=dict(linewidth=3)
    )
plt.legend(labels = ["Non-Factual", "Factual"])
plt.xlabel("")
plt.ylabel("")
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.locator_params(axis='y', nbins=4)
plt.locator_params(axis='x', nbins=5)
plt.xlim(10, 30)
```
